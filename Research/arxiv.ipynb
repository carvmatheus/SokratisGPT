#### Extrator dos BIBlex
Essa parte iremos extrair os biblex e o id do artigo dentro da busca do Arxiv

```python
from bs4 import BeautifulSoup
import requests
import re
import xml.etree.ElementTree as ET

# Loop para avançar de 10 em 10 de 20 até 200
for startat in range(20, 200, 10):
    # Modificar a URL para incluir o valor atual de startat
    url = f'https://search.arxiv.org/?query=%28%22Socratic+method%22+OR+%22Socratic+questioning%22+OR+%22Socratic+dialogue%22+OR+%22Socratic+approach%22+OR+%22Socratic+technique%22+OR+%22Socratic+teaching%22+OR+%22philosophical+inquiry%22%29+AND+%28%22Application%22+OR+%22Software%22%29&qid=1713409579046ler_nCnN_-341835059&byDate=1&startat={startat}'

    response = requests.get(url)
    html = response.text

    soup = BeautifulSoup(html, 'html.parser')

    # Extrair texto do HTML
    text = soup.get_text()

    # Usar expressões regulares para encontrar todos os termos após /abs/
    pattern = r'/abs/([0-9.]+)'
    matches = re.findall(pattern, text)

    print(matches)

    #Construa a string de IDs para a URL
    ids_string = ','.join(matches)

    # Estrutura da URL da API para buscar por IDs específicos
    url = f"http://export.arxiv.org/api/query?id_list={ids_string}&max_results=20"

    # Faça a requisição à API
    try:
        with urllib.request.urlopen(url) as response:
            # Verifique se a requisição foi bem-sucedida
            if response.status == 200:
                # Leia e analise o XML de resposta
                xml_data = response.read()
                root = ET.fromstring(xml_data)
                
                # Namespace do Atom feed
                ns = {'atom': 'http://www.w3.org/2005/Atom'}
                
                # Extraia os detalhes dos artigos
                for entry in root.findall('atom:entry', ns):
                    arxiv_id = entry.find('atom:id', ns).text.split('/')[-1]
                    title = entry.find('atom:title', ns).text.strip()
                    authors_list = [author.find('atom:name', ns).text for author in entry.findall('atom:author', ns)]
                    authors = ', '.join(authors_list)
                    published = entry.find('atom:published', ns).text.split('-')[0]  # Ano de publicação
                    
                    # Imprime a entrada no formato BibTeX
                    print(f'@article{{{arxiv_id.replace("/", "_")},')
                    print(f'  title = {{{title}}},')
                    print(f'  author = {{{authors}}},')
                    print(f'  journal = {{arXiv preprint arXiv:{arxiv_id}}},')
                    print(f'  year = {{{published}}},')
                    print('}\n')

            else:
                print('Falha na requisição à API do arXiv')
    except urllib.error.URLError as e:
        print(f'Falha na requisição à API do arXiv: {e.reason}')
```
